```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import requests
import pandas as pd
```

```python
URL = 'https://data-class-mars-challenge.s3.amazonaws.com/Mars/index.html'
lists = pd.read_html(URL)
```

```python
df = lists[0]
```

```python
df['terrestrial_date']=pd.to_datetime(df['terrestrial_date'])
```

```python
df.groupby(pd.Grouper(key='terrestrial_date', freq='M')).count()
```

There are six months on mars. Data was collected for 35 martian months.

```python
df['m_yr'] = pd.to_datetime(df['terrestrial_date']).dt.to_period('M')
df['year'] = pd.to_datetime(df['terrestrial_date']).dt.to_period('Y')
```

```python
df["min_temp"].count()
```

There are 1867 Days of data

```python

temp_avg = df.groupby(["m_yr"])["min_temp"].mean()

temp_avg_df = pd.DataFrame(temp_avg)

reset = temp_avg_df.reset_index()


chart = reset.plot(kind="bar", x="m_yr", y="min_temp", rot=90, figsize=(15,5),ylim=(-100,-50), grid=True,title= "Temp on Mars by Month")
chart.set_ylabel("Celsius")
```

The highest temp on mars is in Jan 2016. The warmest is in Jan 2013.

```python
atmosphere_avg = df.groupby(["m_yr"])["pressure"].mean()

atmosphere_avg_df = pd.DataFrame(atmosphere_avg)

reset = atmosphere_avg_df.reset_index()

chart = reset.plot(kind="bar", x="m_yr", y="pressure", rot=90, figsize=(15,5), ylim=(500,1000), grid=True,title= "Pressure on Mars by Month")
chart.set_ylabel("Atmospheric Pressue")
```

The lowest atmospheric pressure is in May 2016. The highest atmospheric pressure is in Dec 2014.

```python
reduced_df= df[["terrestrial_date","month","min_temp"]]


chart = reduced_df.plot(kind="line", x="terrestrial_date", y="min_temp", rot=90, figsize=(15,5), grid=True,title= "Temp on Mars by Day")
chart.set_ylabel("Celsius")
chart.set_xlabel("Earth Year")
```

There are nearly 700 days in a martian year.
